```python
import re

def cleanresume(txt):
    cleantxt = re.sub('http\S+', ' ', txt)  # Fixed URL pattern
    cleantxt = re.sub('RT|cc', ' ', cleantxt)
    cleantxt = re.sub('@\S+', ' ', cleantxt)
    cleantxt = re.sub('#\S+', ' ', cleantxt)
    cleantxt = re.sub('[%s]' % re.escape('''!"#$%^&*()*+,-./:;<=>?@[\]^_`{|}~'''), ' ', cleantxt)
    cleantxt = re.sub(r'[^\x00-\x7f]', ' ', cleantxt)
    cleantxt = re.sub('\s+', ' ', cleantxt)  # Fixed: collapse multiple spaces
    return cleantxt.strip()  # Added strip() to remove leading/trailing spaces
```

```python
df['Resume'] = df['Resume'].apply(lambda x:cleanresume(x))
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

```python
df['Category'].unique()
```

```python
# vectorization 

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
reuire_text = tfidf.transform(df['Resume'])
```

```python
reuire_text
```

```python
# train_test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(reuire_text,df['Category'], test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# model training

```python
from sklearn.neighbors import KNeighborsClassifier

from sklearn.multiclass import OneVsRestClassifier

from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print(accuracy_score(y_test, y_pred))
```
